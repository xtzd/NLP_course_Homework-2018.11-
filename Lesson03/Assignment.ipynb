```python
import pandas as pd
```

```python
subway_info = pd.read_csv('Subway_Data-master/北京地铁/time（北京）.csv',engine='python',encoding='utf8')
```

```python
routine_set = set(subway_info['line_no'].tolist())
```

```python
routine={}
for i in routine_set:
    routine[i] = set(subway_info[subway_info['line_no']=='1']['from_station'].apply(lambda x: x.strip(' ')).tolist())
```

```python
routine
```

```python
def transfer_method(from_station, to_station, all_path=True):
    seen = set()
    pathes = [[from_station]]
    chosen_pathes = []
    while pathes:
        path = pathes.pop(0)
        frontier = path[-1]
        if (not all_path) and (frontier in seen): continue
        seen.add(frontier)
        for station in routine[frontier]:
            if station in path: continue
            new_path = path +[station]
            pathes.append(new_path)
            if station == to_station:
                chosen_pathes.append(new_path)
    return chosen_pathes
```

```python
transfer_method('公主坟','北大东门')
```
