```python
import codecs
with codecs.open("10k_articles.txt",'r',encoding="utf-8") as file:
    content = file.readlines()[:10000]
```

```python
content = ''.join(content)
len(content)
```

```python
import re
def tokenize(string):
    return ''.join(re.findall('[\u4e00-\u9fa5]+',string))
```

```python
tokenize(content)[:1000]
```

```python
all_character = tokenize(content)
len(all_character)
```

```python
from collections import Counter

Counter(all_character).most_common()
```

```python
my_counter = Counter(all_character)
occurances = sum(my_counter.values())
def get_prob_from_counts(char):
    return my_counter[char]/occurances
```

```python
get_prob_from_counts('的')
```

```python
from functools import reduce
from operator import mul, add
```

```python
def prob_of_string(string):
    return reduce(mul,[get_prob_from_counts(char) for char in string])
```

```python
prob_of_string("今天我们一起春游")
```

# 2-Gram

```python
gram_len = 2
two_gram_counts = Counter(all_character[i:i+gram_len] for i in range(len(all_character)-gram_len))
```

```python
two_gram_counts.most_common()
```

```python
two_gram_occurances = sum(two_gram_counts.values())
def get_pair_prob_from_counts(two_gram):
    return two_gram_counts[two_gram]/two_gram_occurances
```

```python
get_pair_prob_from_counts('我们')
```

```python
def get_pair_prob_from_2gram(prev_c, c):
    if get_pair_prob_from_counts(prev_c + c)>0:
        return get_pair_prob_from_counts(prev_c + c)
    else:
        return get_prob_from_counts(c)
```

```python
def get_sentence_prob_from_2gram(sentence):
    prob = []
    for i, c in enumerate(sentence):
        prev_c = '<s>' if i==0 else sentence[i-1]
        prob.append(get_pair_prob_from_2gram(prev_c,c))
    return reduce(mul,prob)
```

```python
get_sentence_prob_from_2gram("前天晚上吃早饭的时候")
```
